```python
#importing pandas as pd
import pandas as pd
  
# Read and store content
# of an excel file 
read_file = pd.read_excel ("Yamunapuram.xlsx")
  
# Write the dataframe object
# into csv file
read_file.to_csv ("Ypm.csv", 
                  index = None,
                  header=True)
    
# read csv file and convert 
# into a dataframe object
#skippingheaders
df = pd.DataFrame(pd.read_csv("Ypm.csv", skiprows=16))


# Remove three columns as index base
df.drop(df.columns[[1,3,4,5,6,7,8]], axis = 1, inplace = True)

  
# show the dataframe
df

# pest = df.iloc[:,0]
# print(pest)


    
```

```python
x = df.iloc[0,0]
i1 = x.rfind('-')
i2 = x.find(' ')

year = int(x[i1+1:i2])
print(year)

i3 = x.find('-')
i4 = x.rfind('-')

month = int(x[i3+1:i4])
print(month)
```

```python
l_df = df.shape[0]
ms= []
ys = []

for i in range(l_df):
    x = df.iloc[i,0]
    i1 = x.rfind('-')
    i2 = x.find(' ')

    year = x[i1+1:i2]
    
    i3 = x.find('-')
    i4 = x.rfind('-')

    month = x[i3+1:i4]
    
    ms.append(month)
    ys.append(year)
    
df["Month"] = ms
df["Year"] = ys
df
```

```python
df.columns
```

```python
l1 = []
l2 = []

for i in range(12):
    if i < 9:
        el = "0" + str(i+1)
    else:
        el = str(i+1)
    l1.append(el)
    
for i in range(2017, 2022):
    l2.append(str(i))
```

```python
dict_temp = {}
for el2 in l2:
    for el1 in l1:
        df_temp1 = df[df["Year"] == el2]
        df_temp2 = df_temp1[df_temp1["Month"] == el1]
        
        x1 = df_temp2.shape[0]
        x2 = 0
        x3 = 0
        for i in range(x1):
            if df_temp2.iat[i, 1] == "None":
                x3 = x3 + 1
                continue
            else:
                x2 = x2 + float(df_temp2.iat[i, 1])
        list_temp = [x1, x2, x3]
        st_temp = "MY-" + el1 + "-"+el2
        dict_temp[st_temp] = list_temp
```

```python
dict_temp
```

```python
key_list = list(dict_temp.keys())
```

```python
for k in key_list:
    if dict_temp[k][0] - dict_temp[k][2] == 0:
        l = dict_temp[k]
        l.append(-1)
        dict_temp[k] = l
        continue
    print(dict_temp[k][0]-dict_temp[k][2])
    a = dict_temp[k][1]/(dict_temp[k][0]-dict_temp[k][2])
    l = dict_temp[k]
    l.append(a)
    dict_temp[k] = l
```

```python
dict_temp
```

```python
#to get the particular month with the year 
final_l0 = key_list[:-5]
final_l1 = []
for f in final_l0:
    f1 = f[3:]
    final_l1.append(f1)
final_l1
```

```python
final_l2 = []
final_l3 = []
for i in final_l0:
    if dict_temp[i][0] == 0:
        new = 0
    else:
        new = round((dict_temp[i][2]/dict_temp[i][0])*100, 2)
    final_l3.append(new)
    if dict_temp[i][3] == -1:
        final_l2.append(0)
    else:
        final_l2.append(dict_temp[i][3])
        
dict_f = {"Dates": final_l1, "PM2.5": final_l2, "Missing Days": final_l3}
df_f = pd.DataFrame(dict_f)
#to get the monthly average concentration of PM 2.5       
df_f.to_csv("Ypm_final.csv",
                index = None,
                header = True)
```
