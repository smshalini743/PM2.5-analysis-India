```python
# importing package
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(20, 18), dpi=80)
import numpy as np
import pandas as pd

def addlabels(x,y):
    for i in range(5*len(x)):
        plt.text(i,y[i],y[i])

filename = "./AV_final.csv"
df_1 = pd.read_csv(filename)
ms = []
for i in range(12):
    if i < 9:
        s = "0" + str(i+1)
    else:
        s = str(i+1)

    ms.append(s)

len_1 = df_1.shape[0]
dict_1 = {}

for m in ms:
    arr1 = []
    arr2 = []
    for i in range(len_1):
        i1 = df_1.iat[i, 0].find('-')
        sub = df_1.iat[i, 0][:i1]
        if m == sub:
            arr1.append(df_1.iat[i, 1])
            arr2.append(df_1.iat[i, 2])
    if len(arr1) != 5:
        arr1.append(0)
        arr2.append(100.0)

    dict_1[m] = [arr1, arr2]

x = ms

y11 = []
y12 = []
y21 = []
y22 = []
y31 = []
y32 = []
y41 = []
y42 = []
y51 = []
y52 = []
y = []

for m in ms:
    y11.append(dict_1[m][0][0])
    y.append(dict_1[m][1][0])

#for sanjay place
#y11[5] = 0
#y[5] = 100

#for nehru nagar
#y11[0] = 0
#y[0] = 100

#for vasundhara
#y11[10] = 0
#y[10] = 100

for m in ms:
    y21.append(dict_1[m][0][1])
    y.append(dict_1[m][1][1])

for m in ms:
    y31.append(dict_1[m][0][2])
    y.append(dict_1[m][1][2])

for m in ms:
    y41.append(dict_1[m][0][3])
    y.append(dict_1[m][1][3])

for m in ms:
    y51.append(dict_1[m][0][4])
    y.append(dict_1[m][1][4])

    
ff = 18
# create data
x = np.arange(12)
# y1 = [34, 56, 12, 89, 67]
# y2 = [12, 56, 78, 45, 90]
# y3 = [14, 23, 45, 25, 89]
width = 0.15
count = 0
# plot data in grouped manner of bar type
bars1 = plt.bar(x-2*width, y11, width, color='cyan')
for bar1 in bars1:
    yval = bar1.get_height()
    plt.text(bar1.get_x(), yval + 5, y[count], rotation=90, fontsize=ff)
    count = count + 1

bars2 = plt.bar(x-width, y21, width, color="blue")
count = 12
for bar2 in bars2:
    yval = bar2.get_height()
    plt.text(bar2.get_x(), yval + 5, y[count], rotation=90, fontsize=ff)
    count = count + 1

bars3 = plt.bar(x, y31, width, color='orange')
count = 24
for bar3 in bars3:
    yval = bar3.get_height()
    plt.text(bar3.get_x(), yval + 5, y[count], rotation=90, fontsize=ff)
    count = count + 1

bars4 = plt.bar(x+width, y41, width, color="red")
count = 36
for bar4 in bars4:
    yval = bar4.get_height()
    plt.text(bar4.get_x(), yval + 5, y[count], rotation=90, fontsize=ff)
    count = count + 1

bars5 = plt.bar(x+2*width, y51, width, color='green')
count = 48
for bar5 in bars5:
    yval = bar5.get_height()
    plt.text(bar5.get_x(), yval + 5, y[count], rotation=90, fontsize=ff)
    count = count + 1
    
# addlabels(x, y)

plt.xticks(x, ms, fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(0, 350)

plt.title("Anand Vihar - Hapur", fontsize=25)
plt.xlabel("Months", fontsize=20)
plt.ylabel("PM2.5", fontsize=20)
legend1 = plt.legend(["2017", "2018", "2019", "2020", "2021"], loc=9, fontsize = ff)
plt.axhline(y = 10, color = 'b', linestyle = ':', label = "WHO")
plt.axhline(y = 40, color = 'r', linestyle = ':', label = "NAAQS")
lines = plt.gca().get_lines()
include = [0,1]
legend2 = plt.legend(["WHO", "NAAQS"], loc=2, fontsize = 16)
plt.gca().add_artist(legend1)

plt.show()
```
